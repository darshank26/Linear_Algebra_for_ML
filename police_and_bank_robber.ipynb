```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,80,2000)
```

```python
d_robber = 2.5 * t
```

```python
d_police = 3 * ( t - 5)
```

```python
fig , ax = plt.subplots()
plt.title('A bank robber caught')
plt.xlabel('Time in minutes')
plt.ylabel('Distance in KM')
ax.set_xlim([0,80])
ax.set_ylim([0,200])
ax.plot(t,d_robber,c='green')
ax.plot(t,d_police,c='red')
plt.axvline(x=30,color='black', linestyle = '--')
plt.axhline(y=75,color='black', linestyle = '--')
```
